```python
from importlib import reload
from my_functions import read_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
elec_data_path = './elec_data/hac.xlsx'
weather_data_path = './weather_data/weather_data_20170101-20170507_Shenzhen.csv'

elec = read_data.elec(elec_data_path, scale='H')
weather = read_data.weather(weather_data_path)
big_data = read_data.merge_elec_and_weather(elec, weather, startDate='1/15/2017', endDate='5/7/2017')
hac = read_data.drop_col(big_data, keep_columns=['kwh', 'tempm', 'hum'])
```

```python
elec_data_path = './elec_data/hhy.xlsx'
weather_data_path = './weather_data/weather_data_20170101-20170507_Shaoguan.csv'

elec = read_data.elec(elec_data_path, scale='H')
weather = read_data.weather(weather_data_path)
big_data = read_data.merge_elec_and_weather(elec, weather, startDate='1/15/2017', endDate='5/7/2017')
hhy = read_data.drop_col(big_data, keep_columns=['kwh', 'tempm', 'hum'])
```

```python
f, ax = plt.subplots(2, sharex=True, figsize=(12, 10))
l1 = ax[0].plot(hhy.index.hour, hhy.kwh, '.', color='grey')
l2 = ax[0].plot(hhy.kwh.groupby(hhy.index.hour).mean())
ax[0].legend((l1[0], l2[0]), ('data point', 'Avg'), loc='upper left', fontsize=12)
ax[0].set_title('Hotel A')

ax[1].plot(hac.index.hour, hac.kwh, '.', color='grey')
ax[1].plot(hac.kwh.groupby(hac.index.hour).mean())
ax[1].set_title('Shopping Mall B')
ax[1].set_xlabel('Hour of day')
ax[1].set_xticks(range(24))

f.text(0.07, 0.5, 'Elec. Usage (kwh)', va='center', rotation='vertical', fontsize=14)
```
